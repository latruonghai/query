```python
from Crawl import Crawl, random_char
import re
from datetime import *
help(Crawl.__init__)
```

```python
class CrawlWiki(Crawl):
    def __init__(self, url):
        super().__init__(url)
    def letCrawl(self):
        soup = self.get_page_content(self.url)
        print(soup)
```

```python
cr =CrawlWiki('https://vi.wikipedia.org/wiki/Wikipedia')
```

```python
a = cr.get_page_content(cr.url)
```

```python
b = a.findAll(('h2','p'))
```

```python
d = []
l = len(b)
i = 0
while True:
    if i == l:
        break
    else:
        kw = b[i].find(class_ = 'mw-headline')
        if kw != None:
            temp = []
            text = kw.text
            text += '\n'
            for j in range(i+1,l):
                #print('j: ',j)
                kw = b[j].find(class_ = 'mw-headline')
                if kw != None:
                    #print('i1: ' ,i)
                    temp.append(kw.text)
                    i = j-1
                    break
                else:
                    text+=b[j].text
            #print('kw.text',text)
            #print('i2: ',i)
            d.append(text)
        else:
            d.append(b[i].text)
        i+=1
print(d)
```
